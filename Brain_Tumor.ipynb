```python
datadir='C:/Users/aksha/Desktop/DL/CNN_brain_project/Brain_Tumor/Training'
```

```python
#pre-processing of training data
x_train=[]
y_train=[]
categories=['glioma','meningioma','notumor','pituitary']
for i in categories:
    path=os.path.join(datadir,i)
    for img in os.listdir(path):
        imgpath=os.path.join(path,img)  
        img_array=imread(imgpath)
        img_resize=resize(img_array,(150,150,1))
        x_train.append(img_resize) #no flatten here as we are doing in CNN
        y_train.append(categories.index(i))  #label encoding
```

```python
datadir1='C:/Users/aksha/Desktop/DL/CNN_brain_project/Brain_Tumor/Testing'
```

```python
#pre-processing of testing data
x_test=[]   
y_test=[]
categories1=['glioma','meningioma','notumor','pituitary']
for i in categories1:
    path1=os.path.join(datadir1,i)
    for img1 in os.listdir(path1):
        imgpath1=os.path.join(path1,img1)  
        img_array1=imread(imgpath1)
        img_resize1=resize(img_array1,(150,150,1))
        x_test.append(img_resize1) #no flatten here as we are doing in CNN
        y_test.append(categories1.index(i))  #label encoding
```

```python
#converting lists into numpy arrays
x_train=np.array(x_train)
x_test=np.array(x_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
#model building
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=(150,150,1),activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Flatten())
model.add(Dense(100,activation='relu'))
model.add(Dense(50,activation='relu'))  
model.add(Dense(4,activation='softmax'))
```

```python
#compile
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#fit the model
model.fit(x_train,y_train,epochs=20,batch_size=50)
```

```python
#model evaluate
model.evaluate(x_test,y_test,batch_size=12)   
```
